```python
def video2frame(video_root,frame_root,numOfFrame,sample_rate=5):
    file_list = os.listdir(video_root)
    i=0
    for videos_name in file_list:
        
        video_path=video_root+videos_name
        frame_path=frame_root+str(i)+'/'
        if not os.path.exists(frame_path):
            os.mkdir(frame_path)
        cap=cv.VideoCapture(video_path)
        cap_len=cap.get(7) 
        
        num=range(0,int(cap_len),(sample_rate))
        itera=0
        for f in num:
            
            cap.set(cv.CAP_PROP_POS_FRAMES, f)
            frame_name=frame_path+str(itera)+'.jpg'
            _,image=cap.read()
            #print(frame_name)
            cv.imwrite(frame_name,image) 
            itera+=1
        
        i+=1
        if i==1:
            break

    #cv.destroyAllWindows()
video_root='C:/Users/jingy/Jupyter/Feature/frames/Arson/Arson/'
frame_root='C:/Users/jingy/Jupyter/Fight/frames/Arson/'
video2frame(video_root=video_root,frame_root=frame_root,numOfFrame=None)
```

```python
def frames2video(fps, save_path, frames_path):
    fourcc = cv.VideoWriter_fourcc(*'mp4v')
    
    first=True
    for i in range(100):
        if os.path.isfile("%s/%d.jpg"%(frames_path, i)):
            #firts
            if first:
                frame = cv.imread("%s/%d.jpg"%(frames_path, i))
                size=(frame.shape)
                videoWriter = cv.VideoWriter(save_path, fourcc, fps, (int(size[0]),int(size[1])))
                first=False
            frame = cv.imread("%s/%d.jpg"%(frames_path, i))
            
            videoWriter.write(frame)
    videoWriter.release()
    return
 
frame_path = "D:/RWF_30frames/train/negative/"
save_path ="D:/RWF_Video/negative/"
file_list = os.listdir(frame_path)
i = 0
for fn in file_list:
    print("prosessing: ",fn)
    frames2video(30,os.path.join(save_path+"negative"+str(i)+".mp4"),
                   os.path.join(frame_path+fn))
    i += 1
    
    
    
```
